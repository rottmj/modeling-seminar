```python
# load weigths
agent.actor.load_weights(r"Experiment 1\best\actor.h5")

# get data
velocities = []
accels = []
gaps = []


env = constantVelocityEnvironment.ConstantVelocityEnvironment()
# env.reset()
observation = np.array(env.get_state())
done = False
score = 0
for j in range(200):
        #print(observation)
        action = agent.choose_action(observation, True)
        action_ = action.numpy()[0]
        observation_, reward, done = env.step(action_)
        #print(observation_, reward, done)
        #print(action_)
        #print(env.get_state())
        velocities.append(env.ego_velocity)
        accels.append(action_)
        gaps.append(env.gap)
        score += reward
        observation = np.array(observation_)
        #print(observation)
        if done:
            break
```

```python
import os
plt.style.use('default')
directory = r'Experiment 5\evolution'
i = 0
plt.style.use('default')
for filename in os.listdir(directory):
    
    f = os.path.join(directory, filename)
    # load weigths
    agent.actor.load_weights(f)

    # get data
    velocities = [25]
    accels = []
    gaps = []

    init_values = data_init[indices[i]]
    
    initial_speed_ego = init_values[2]
    initial_speed_lead = init_values[3]
    initial_gap = init_values[0]
    initial_safety_gap = init_values[1]
    
    print(filename)
    print(indices[i])
    
    env = constantVelocityEnvironment.ConstantVelocityEnvironment(initial_speed_ego, initial_speed_lead, initial_gap, initial_safety_gap)
    # env.reset()
    observation = np.array(env.get_state())
    done = False
    score = 0
    for j in range(200):
            #print(observation)
            action = agent.choose_action(observation, True)
            action_ = action.numpy()[0]
            observation_, reward, done = env.step(action_)
            #print(observation_, reward, done)
            #print(action_)
            #print(env.get_state())
            velocities.append(env.ego_velocity)
            accels.append(action_)
            gaps.append(env.gap)
            score += reward
            observation = np.array(observation_)
            #print(observation)
            if done:
                break
    plt.plot(velocities, color="blue")
    plt.xlabel("t [1/10s]")
    plt.ylabel("velocity [m/s]")

    plt.savefig(str(indices[i]) + '.png', dpi=1000)
    plt.close()
    i += 1
```
